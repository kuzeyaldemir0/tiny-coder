### Function for counting frequency of every adjacent pair

```python
from collections import defaultdict

sample_string = "aaabbaababbaaabba"
sample_encoded = sample_string.encode("utf-8")

def pair_freq_counter(ids: list[int]) -> dict[tuple[int, int], int]:
    freq_counter = defaultdict(int)
    for idx in range(len(ids) - 1):
        curr = ids[idx]
        next = ids[idx + 1]
        freq_counter[curr, next] += 1
    return freq_counter

sample_freq_counter = pair_freq_counter(sample_encoded)
print(list(sample_encoded))
print(sample_freq_counter)
```

```python
from collections.abc import Mapping

Pair = tuple[int, int]

def select_pair(
        freq_counter: Mapping[Pair, int],
        min_frequency: int = 2
) -> Pair | None:
    most_freq_pair = max(freq_counter.keys(), key=lambda x:freq_counter[x])
    if freq_counter[most_freq_pair] < min_frequency:
        return None
    return most_freq_pair

most_freq_pair = select_pair(sample_freq_counter)
print(list(sample_encoded))
print("Most frequent pair is:", most_freq_pair)
```

```python
def apply_merge(
        ids: list[int],
        pair_to_merge: tuple[int, int],
        new_id: int
) -> list[int]:
    
    new_ids = []
    idx = 0
    # Manually loop over all adjacent pairs and apply merge if applicable
    while idx + 1 < len(ids):
        curr = ids[idx]
        next_token = ids[idx + 1]
        if curr == pair_to_merge[0] and next_token == pair_to_merge[1]:
            new_ids.append(new_id)
            idx += 2
        else:
            new_ids.append(curr)
            idx += 1
    # Add the last token to the new sequence
    if idx < len(ids):
        new_ids.append(ids[idx])
    
    return new_ids

merged_sample = apply_merge(sample_encoded, most_freq_pair, 256)
print(sample_string)
print("Before merging:")
print(list(sample_encoded))
print("After one merge")
print(merged_sample)
```
